```python
import seaborn as sns

data=sns.load_dataset("titanic")
data
```

```python
data.sibsp.value_counts()
```

```python
data.parch.value_counts()
```

```python

data=data.rename(columns={"sex":"Gender"})
data
```

```python
data.value_counts()
```

```python
data.isnull().sum()
```

```python
mean_age=data.age.mean()
data.age=data.age.fillna(mean_age)
```

```python
data.deck.nunique()
```

```python
data.deck.value_counts()
```

```python
data.deck=data.deck.fillna(method="bfill")
data
```

```python
data.deck=data.deck.fillna(method="pad")
data
```

```python
data=data.drop("embarked",axis=1)
data
```

```python
mode_town=data.embark_town.mode()[0]
data.embark_town=data.embark_town.fillna(mode_town)
data
```

```python
data.isnull().sum()
```
